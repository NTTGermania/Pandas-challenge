```python
# Dependencies and Setup
import pandas as pd
```

```python
# File to Load
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data_df = pd.read_csv(school_data_to_load)
student_data_df = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete_df = pd.merge(student_data_df, school_data_df, how="left", on=["school_name", "school_name"])
school_data_complete_df.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# Calculate the number of schools
schools_unique_df = school_data_complete_df["school_name"].unique()
total_schools_unique = school_data_complete_df["school_name"].nunique()
total_schools_unique
```

```python
# Calculate the number of students
total_students = school_data_complete_df["student_name"].size
total_students
```

```python
# Calculate total budgets
school_budget_summary = school_data_df["budget"].sum()
school_budget_summary
```

```python
# Calculate the average maths score for district summary
average_maths_score = school_data_complete_df["math_score"].mean()
average_maths_score
```

```python
# Calculate the average reading score for district summary
average_reading_score = school_data_complete_df["reading_score"].mean()
average_reading_score
```

```python
# Calculate the percentage of students in the district that have 70 or higher on their maths scores
passing_maths_total_df = school_data_complete_df.loc[school_data_complete_df["math_score"] >= 70,:]
passing_maths_total_students = passing_maths_total_df["student_name"].size
passing_maths_percentage = passing_maths_total_students/ total_students * 100
passing_maths_percentage
```

```python
# Calculate the percentage of students in the district that have 70 or higher on their reading scores
passing_reading_total_df = school_data_complete_df.loc[school_data_complete_df["reading_score"] >= 70,:]
passing_reading_total_students = passing_reading_total_df["student_name"].size
passing_reading_percentage = passing_reading_total_students/ total_students * 100
passing_reading_percentage
```

```python
# Calculate the percentage of students in the distrcict that have 70 or higher on BOTH their maths and reading scores
passing_both_total_df = school_data_complete_df.loc[(school_data_complete_df["math_score"] >= 70) & (school_data_complete_df["reading_score"] >= 70) ,:]
passing_both_total_students = passing_both_total_df["student_name"].size
passing_both_percentage = passing_both_total_students/ total_students * 100
passing_both_percentage
```

```python
# Creating a new dataframe that contains the district summary data
district_summary_df = pd.DataFrame([[total_schools_unique,total_students,school_budget_summary,average_maths_score,average_reading_score,passing_maths_percentage,passing_reading_percentage,passing_both_percentage]], columns = ['Total Schools','Total Students','Total Budget','Average Maths Score','Average Reading Score','% Passing Maths','% Passing Reading','% Overall Passing'])
district_summary_df["Total Students"] = district_summary_df["Total Students"].astype(int).map("{:,}".format)
district_summary_df["Total Budget"] = district_summary_df["Total Budget"].astype(int).map("${:,.2f}".format)
district_summary_df
```

```python
# Dropping "School ID" column from the newly merged dataframe
group_by_schools_merge_df = group_by_school_merge_df.drop(columns='School ID')
group_by_schools_merge_df
```

```python
# Extracting the necessary column from the per_student_budget dataframe
extracted_col = per_student_budget["Per Student Budget"]
```

```python
# Joining the "Per Student Budget" column with the group by schools dataframe
group_by_schools_merge_df = group_by_schools_merge_df.join(extracted_col)
group_by_schools_merge_df
```

```python
# Count the total number of students in each school
total_students_per_school = student_data_df.groupby("school_name").count()
total_students_per_school
```

```python
# Extracting a column with the total number of students per school
total_students_per_school = total_students_per_school["Student ID"]
total_students_per_school
```

```python
# Filtering the school data complete dataframe by only students who got 70 or higher on their maths scores 
school_data_complete_passing_maths = school_data_complete_df.loc[school_data_complete_df["math_score"] >= 70,:]
school_data_complete_passing_maths
# Count the number of students who got 70 or higher on their maths scores per school 
group_by_passing_maths = school_data_complete_passing_maths.groupby("school_name").count()
group_by_passing_maths
# Extracting a column with the number of students who got 70 or higher on their maths scores per school
group_by_passing_maths = group_by_passing_maths ["Student ID"] 
group_by_passing_maths
# Calculate the percentage of students who got 70 or higher on their maths scores per school
percent_passing_maths_df = group_by_passing_maths / total_students_per_school * 100
percent_passing_maths_df
```

```python
# Filtering the school data complete dataframe by only students who got 70 or higher on their reading scores 
school_data_complete_df_passing_reading = school_data_complete_df.loc[school_data_complete_df["reading_score"] >= 70,:]
school_data_complete_df_passing_reading
# Count the number of students who got 70 or higher on their reading scores per school 
group_by_passing_reading = school_data_complete_df_passing_reading.groupby("school_name").count()
group_by_passing_reading
# Extracting a column with the number of students who got 70 or higher on their reading scores per school
group_by_passing_reading = group_by_passing_reading["Student ID"]
group_by_passing_reading
# Calculate the percentage of students who got 70 or higher on their reading scores per school
percent_passing_reading_df = group_by_passing_reading / total_students_per_school * 100
percent_passing_reading_df
```

```python
# Filtering the school data complete dataframe by only students who got 70 or higher on both their maths and reading scores 
school_data_complete_df_passing_overall = school_data_complete_df.loc[(school_data_complete_df["reading_score"] >= 70) & (school_data_complete_df["math_score"]>= 70),:]
school_data_complete_df_passing_overall
# Count the number of students who got 70 or higher on both their maths and reading scores per school
group_by_passing_overall = school_data_complete_df_passing_overall.groupby("school_name").count()
group_by_passing_overall
# Extracting a column with the number of students who got 70 or higher on both their maths and reading scores per school
group_by_passing_overall = group_by_passing_overall["Student ID"]
group_by_passing_overall
# Calculate the percentage of students who got 70 or higher on both their maths and reading scores per school
percent_passing_overall_df = group_by_passing_overall / total_students_per_school * 100
percent_passing_overall_df
```

```python
# Creating a school summary dataframe by merging the group_by_schools dataframe with the percent passing maths dataframe
school_summary_df = pd.merge(group_by_schools_merge_df,percent_passing_maths_df, on = "school_name", how = "outer" )
# Renaming the column to its appropriate name
school_summary_df = school_summary_df.rename(columns = {"Student ID" : "% Passing Math"})
school_summary_df
```

```python
# Merging the percentage of passing reading and maths into the school summary dataframe
school_summary_df = pd.merge(school_summary_df,percent_passing_overall_df, on = "school_name", how = "outer" )
# Renaming the column to its appropriate name
school_summary_df = school_summary_df.rename(columns = {"Student ID" : "% Overall Passing"})
```

  * % Passing Math

```python
school_spending_df = school_summary_df[["Per Student Budget", "Average Math Score", "Average Reading Score", "% Passing Math", "% Passing Reading", "% Overall Passing"]]
school_spending_df
```

```python
# Create bins for spending ranges per student
bins_spending = [0, 585, 630, 645, 680]
bins_spending_labels = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
# Putting budget per student intheir respective bins
school_spending_df["Spending Ranges (Per Student)"] = pd.cut(school_spending_df["Per Student Budget"], bins_spending, labels = bins_spending_labels)
school_spending_df
```

```python
# Grouping the data by the spending ranges per student
school_spending_final_df = school_spending_df.groupby("Spending Ranges (Per Student)").mean()
```

```python
# Formatting the dataframe
school_spending_final_df = school_spending_final_df.round(2)
school_spending_final_final_df = school_spending_final_df[["Average Math Score","Average Reading Score","% Passing Math" , "% Passing Reading" , "% Overall Passing"]]
# Final version of the dataframe
school_spending_final_final_df
```
